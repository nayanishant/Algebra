```python
import matplotlib.pyplot as plt
import numpy as np

Xmin = -10
Ymin = -10
Xmax = 10
Ymax = 10

# Define how many points to plot
points = 10*(Xmax-Xmin)

# Define the arraay of x values once
x = np.linspace(Xmin, Xmax, points)

fig, ax = plt.subplots()
plt.axis([Xmin, Xmax, Ymin, Ymax])
plt.plot([Xmin, Xmax], [0, 0], 'b')
plt.plot([0, 0], [Ymin, Ymax], 'b')

# Line 1
y1 = 2*x
plt.plot(x, y1)

# Line 2
y2 = x**2 - 3
plt.plot(x, y2)

plt.show()
```
