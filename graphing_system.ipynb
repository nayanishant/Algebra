Shading and graphing inequalities

```python
import matplotlib.pyplot as plt
import numpy as np

Xmin = -10
Ymin = -10
Xmax = 10
Ymax = 10

# Define how many points to plot
points = 10*(Xmax-Xmin)

# Define the arraay of x values once
x = np.linspace(Xmin, Xmax, points)

fig, ax = plt.subplots()
plt.axis([Xmin, Xmax, Ymin, Ymax])
plt.plot([Xmin, Xmax], [0, 0], 'b')
plt.plot([0, 0], [Ymin, Ymax], 'b')

# Line 1
y1 = x + 6
plt.plot(x, y1)
plt.fill_between(x, y1, 10, facecolor='red')

# Line 2
y2 = x + 3
plt.plot(x, y2)
plt.fill_between(x, y2, y1, facecolor='yellow')

# Line 3
y3 = x - 1
plt.plot(x, y3)
plt.fill_between(x, y3, y2, facecolor='green')

# Line 4
y4 = x - 4
plt.plot(x, y4)
plt.fill_between(x, y4, y3, facecolor='blue')

plt.show()
```

Interactive graph

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b, zoom):
    Xmin = -zoom
    Ymin = -zoom
    Xmax = zoom
    Ymax = zoom
    points = 10*(Xmax-Xmin)
    x = np.linspace(Xmin, Xmax, points)

    plt.axis([Xmin, Xmax, Ymin, Ymax])
    plt.plot([Xmin, Xmax], [0, 0], 'black')
    plt.plot([0, 0], [Ymin, Ymax], 'black')
    plt.title('y = mx + b')

    plt.plot(x, m*x+b)
    plt.show()

# Set up the sliders
interactive_plot = interactive(f, m = (-9, 9), b = (-9, 9), zoom = (1, 100))
interactive_plot
```

```python
# Graphing weather data

from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily

# set time period
start = datetime(2024, 2, 1)
end = datetime(2024, 2, 25)

mandya = Point(12.522157, 76.900917)

data = Daily(mandya, start, end)
data = data.fetch()

# Plot line chart including average, minimum and maximum temperature
data.plot(y = ['tavg', 'tmin', 'tmax'])
plt.show()
```
