Question:
A town's population increases at a constant rate. In 2010 the population was 55,000. By 2012 th population increased to 76,000. If this trend continues, predict the population in 2016.

```python
import matplotlib.pyplot as plt

x1 = 0
x2 = 2
y1 = 55
y2 = 76

m = (y2 - y1)/(x2 - x1)
b = y1 - m*x1
print('y = ', m, 'x + ', b)

# For graph
xMin = 0
yMin = 0
xMax = 10
yMax = 150

y3 = m*xMin + b
y4 = m*xMax + b

fig, ax = plt.subplots()
plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

# Add details to the graph
ax.set_xlabel("time")
ax.set_ylabel("population")
ax.grid(True)

# Plot the linear function as a red line
plt.plot([xMin, xMax], [y3, y4], 'r')

plt.show()
```

Question:
The number of people afflicted with the common cold in winter months dropped steadily by 50 each year since 2004 until 2010. In 2004, 875 people were inflicted. Find the linear function that models the number of people afflicted with the common cold C as a function of the year, t. When will no one be afflicted?

```python
import matplotlib.pyplot as plt
import sympy
from sympy import symbols
from sympy.solvers import solve

x1 = 0
x2 = 1
y1 = 875
y2 = 825

m = (y2 - y1)/(x2 - x1)
b = y1 - m*x1
x = symbols('x')
print('y = ', m, 'x + ', b)
eq = 875 -50*x
sol = solve(eq, x)
print('x = ', sol[0])

# For graph
xMin = 0
yMin = 0
xMax = 20
yMax = 900

y3 = m*xMin + b
y4 = m*xMax + b

fig, ax = plt.subplots()
plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

# Add details to the graph
ax.set_xlabel("time")
ax.set_ylabel("number of people")
ax.grid(True)

# Plot the linear function as a red line
plt.plot([xMin, xMax], [y3, y4], 'r')

plt.show()
```
