```python
# Simplest way to solve a system of equations. Set each equation equal to 0

from sympy import *
x, y = symbols('x y')

# Equation set equal to zero
first = 2*x + y - 1
second = x - 2*y + 7

# Solution
solution = linsolve([first, second], (x, y))
x_sol = solution.args[0][0]
y_sol = solution.args[0][1]

print("Solution = (", x_sol, ",", y_sol, ")")
```

```python
# Solution with graph
from sympy import *
x, y = symbols('x y')

# Equation set equal to zero
first = 2*x + y - 1
second = x - 2*y + 7

# Solution
solution = linsolve([first, second], (x, y))
x_sol = solution.args[0][0]
y_sol = solution.args[0][1]

print("Solution = (", x_sol, ",", y_sol, ")")

# Sympy syntax for equation equal to zero, ready to factor
y_first = Eq(first, 0)

# Sympy solve for y
y1 = solve(y_first, y)

# Same steps for second equation
y_second = Eq(second, 0)
y2 = solve(y_second, y)

# Show factorial result
print("y = ", y1[0])
print("y = ", y2[0])

# Plot solution
x = symbols('x')
plot(y1[0], y2[0], (x, -10, 10))
```
