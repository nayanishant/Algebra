Function and a table of (x,y) values

```python
x = 5
y = 4*x + 3 #The function

print(x,",",y)

print("x \t y")

for x in range(11):
    y = 4*x + 3
    print(x, "\t", y)
```

Defining a Python function

```python
def f(x):
    y = 4*x + 3
    return y

print(5, ",", f(5))

for x in range(11):
    print(x, "\t", f(x))
```

```python
# Display axis line
import matplotlib.pyplot as plt

xMin = -10
xMax = 10
yMin = -10
yMax = 10

fig, ax = plt.subplots()

plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

plt.show()
```

```python
# Plot a point
import matplotlib.pyplot as plt

xMin = -10
xMax = 10
yMin = -10
yMax = 10

fig, ax = plt.subplots()

plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

plt.plot([5], [3], 'ro')

plt.show()
```

```python
# Plot several points
import matplotlib.pyplot as plt

xMin = -10
xMax = 10
yMin = -10
yMax = 10

fig, ax = plt.subplots()

plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

for x in range(10):
    y = 0.5*x + 1
    plt.plot([x], [y], 'ro')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xMin = -10
xMax = 10
yMin = -10
yMax = 10
points = 2*(xMax-xMin)
x = np.linspace(xMin, xMax, points)

fig, ax = plt.subplots()

plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

y = 2*x + 1
plt.plot(x, y, 'r')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xMin = -10
xMax = 10
yMin = -10
yMax = 10
points = 2*(xMax-xMin)
x = np.linspace(xMin, xMax, points)

fig, ax = plt.subplots()

plt.axis([xMin, xMax, yMin, yMax]) # Window size
plt.plot([xMin, xMax], [0, 0], 'b') # blue x axis
plt.plot([0, 0], [yMin, yMax], 'b') # blue y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.grid(True)

ax.set_xticks(np.arange(xMin, xMax, 2))
ax.set_yticks(np.arange(yMin, yMax, 2))

y = 2*x + 1
plt.plot(x, y, label = "y=2x+1")
plt.plot([4], [6], 'ro', label = 'points')
plt.plot(x, 3*x, label = 'steeper line')
plt.legend()
plt.show()
```
